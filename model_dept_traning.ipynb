```python
import fiftyone as fo
import fiftyone.zoo as foz
import fiftyone.utils.random as forand
from ultralytics import YOLO
from PIL import Image
from transformers import pipeline
import torch
import os

# Load Datatsets
dataset_oiv7 = foz.load_zoo_dataset(
    "open-images-v7",
    split="train",
    label_types=["detections"],
    classes=classes_oiv7,
    max_samples=3000,
)

dataset_coco = foz.load_zoo_dataset(
    "coco-2017",
    split="train",
    label_types=["detections"],
    classes=classes_coco,
    max_samples=3000,
)

# Clean Dataset
for dataset in [dataset_coco, dataset_oiv7]:
    to_remove = [s.id for s in dataset if not os.path.exists(s.filepath)]
    dataset.delete_samples(to_remove)

# Merge and Split datasets (prep)
dataset_coco.merge_samples(dataset_oiv7)
forand.random_split(dataset_coco, {"train": 0.8, "val": 0.2})

# YOLO Formatting
for split in ["train", "val"]:
    dataset_coco.match_tags(split).export(
        export_dir="yolo_data",
        dataset_type=fo.types.YOLOv5Dataset,
        split=split,
        classes=classes_coco + classes_oiv7,
    )


# Fine Tune YOLO26n model
model = YOLO("yolo26n.pt")
model.train(
    data="yolo_data/dataset.yaml",
    epochs=10,
    imgsz=640,
    device=0,
)
```

```python

#Dept Calcualtion
depth_estimator = pipeline(
    task="depth-estimation",
    model="depth-anything/Depth-Anything-V2-Small-hf"
)

test_image = Image.open("testing-imgs/room0.png")
depth_output = depth_estimator(test_image)
depth_map = depth_output["depth"]
depth_map.save("testing-results/depth_map.png")

# Detection + Depth
results = model("testing-imgs/room0.png", conf=0.05)

for i, r in enumerate(results):
    im_bgr = r.plot()
    im_rgb = Image.fromarray(im_bgr[..., ::-1])

    combined = Image.new("RGB", (im_rgb.width + depth_map.width, im_rgb.height))
    combined.paste(im_rgb, (0, 0))
    combined.paste(depth_map.resize((depth_map.width, im_rgb.height)), (im_rgb.width, 0))
    combined.save(f"testing-results/results_with_depth{i}.jpg")
```

```python
from transformers import pipeline
from PIL import Image
import os

os.makedirs("testing-results", exist_ok=True)

image_path = "room0.png"

# Depth
depth_estimator = pipeline(
    task="depth-estimation",
    model="depth-anything/Depth-Anything-V2-Small-hf"
)

test_image = Image.open(image_path).convert("RGB")

depth_output = depth_estimator(test_image)
depth_map = depth_output["depth"].convert("RGB")

depth_map.save("testing-results/depth_map.png")

# Detection + Depth
results = model(image_path, conf=0.05)

for i, r in enumerate(results):
    im_bgr = r.plot()
    im_rgb = Image.fromarray(im_bgr[..., ::-1]).convert("RGB")

    depth_resized = depth_map.resize((im_rgb.width, im_rgb.height))

    combined = Image.new("RGB", (im_rgb.width * 2, im_rgb.height))
    combined.paste(im_rgb, (0, 0))
    combined.paste(depth_resized, (im_rgb.width, 0))

    combined.save(f"testing-results/results_with_depth{i}.jpg")
```

```python
from transformers import pipeline
from PIL import Image
import numpy as np
import cv2
import os

os.makedirs("testing-results", exist_ok=True)

image_path = "blue_bedroom.jpg"

# Depth model
depth_estimator = pipeline(
    task="depth-estimation",
    model="depth-anything/Depth-Anything-V2-Small-hf"
)

# Load image
test_image = Image.open(image_path).convert("RGB")

# Run depth estimation
depth_output = depth_estimator(test_image)

# Raw depth array
depth_array = np.array(depth_output["depth"])

# Normalize depth values to 0-255
depth_normalized = cv2.normalize(
    depth_array,
    None,
    0,
    255,
    cv2.NORM_MINMAX
).astype(np.uint8)

# Apply color map
# COLORMAP_JET:
# Red = near
# Blue = far
depth_colormap = cv2.applyColorMap(
    depth_normalized,
    cv2.COLORMAP_JET
)

# Convert BGR -> RGB
depth_colormap_rgb = cv2.cvtColor(
    depth_colormap,
    cv2.COLOR_BGR2RGB
)

# Convert to PIL image
depth_map = Image.fromarray(depth_colormap_rgb)

# Save colored depth map
depth_map.save("testing-results/blue_bedroom_depth_colored.png")
```
