## Training COCO-pt model on YOLO data

```python
import fiftyone as fo
import fiftyone.zoo as foz
import fiftyone.utils.random as forand
import os

dataset_oiv7 = foz.load_zoo_dataset(
    "open-images-v7",
    split="train",
    label_types=["detections"],
    classes=classes_oiv7,
    max_samples=500
)

dataset_coco = foz.load_zoo_dataset(
    "coco-2017",
    split="train",
    label_types=["detections"],
    classes=classes_coco,
    max_samples=500
)

to_remove=[]
for i in dataset_coco:
    if not os.path.exists(i.filepath):
        to_remove.append(i.id)
dataset_coco.delete_samples(to_remove)

to_remove=[]

for i in dataset_oiv7:
    if not os.path.exists(i.filepath):
        to_remove.append(i.id)
dataset_oiv7.delete_samples(to_remove)


# forand.random_split(dataset_oiv7, {"new_train": 0.8, "new_val": 0.2})
forand.random_split(dataset_coco, {"train": 0.8, "val": 0.2})


# dataset_train.export(
#     export_dir='dataset/train',
#     dataset_type=fo.types.YOLOv5Dataset,
#     label_field='ground_truth'
# )

for split in ["train", "val"]:
    split_view = dataset_coco.match_tags(split)
    split_view.export(
        export_dir='yolo_data',
        dataset_type=fo.types.YOLOv5Dataset,
        split=split,
        classes=classes_coco+classes_oiv7,
        label_field='ground_truth'
    )
```
