## Training COCO-pt model on YOLO data

```python
import fiftyone as fo
import fiftyone.zoo as foz
import fiftyone.utils.random as forand


splits = ['train','validation']
classes=["Bathroom cabinet", "Bench"]


dataset_train = foz.load_zoo_dataset(
    "open-images-v7",
    split="train",
    label_types=["detections"],
    classes=classes,
    max_samples=1000,
)

forand.random_split(dataset_train, {"train": 0.8, "val": 0.2})


classes = dataset_train.distinct("ground_truth.detections.label")

# dataset_train.export(
#     export_dir='dataset/train',
#     dataset_type=fo.types.YOLOv5Dataset,
#     label_field='ground_truth'
# )

for split in ["train", "val"]:
    split_view = dataset_train.match_tags(split)
    split_view.export(
        export_dir='yolo_data',
        dataset_type=fo.types.YOLOv5Dataset,
        label_field='ground_truth',
        split=split,
        classes=classes,
    )
```

```python
from ultralytics import YOLO

# COCO-pretrained YOLO26n model
model = YOLO("yolo26n.pt")

results = model.train(data="yolo_dataset/dataset.yaml", epochs=100, imgsz=640)
```
