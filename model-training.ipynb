## Training COCO-pt model on YOLO data

```python
import fiftyone as fo
import fiftyone.zoo as foz
import fiftyone.utils.random as forand
import os

dataset_oiv7 = foz.load_zoo_dataset(
    "open-images-v7",
    split="train",
    label_types=["detections"],
    classes=classes_oiv7,
    max_samples=5000,
    only_matching=True,
)


to_remove=[]

for i in dataset_oiv7:
    if not os.path.exists(i.filepath):
        to_remove.append(i.id)
dataset_oiv7.delete_samples(to_remove)
print('----',to_remove)


forand.random_split(dataset_oiv7, {"train": 0.8, "val": 0.2})


for split in ["train", "val"]:
    split_view = dataset_oiv7.match_tags(split)
    split_view.export(
        export_dir='yolo_data',
        dataset_type=fo.types.YOLOv5Dataset,
        split=split,
        classes=classes_oiv7,
        label_field='ground_truth'
    )
```

```python
model = YOLO("yolo26n.pt")

results = model.train(data="yolo_data/dataset.yaml", epochs=30, imgsz=640)

# model = YOLO("/Users/elena/Documents/GitHub/model-code/yolo26n.pt")
# results = model.train(data="yolo_data/dataset.yaml", epochs=100, imgsz=640)

# training done in colab notebook!
# https://colab.research.google.com/drive/1jQFEuRHpO4FJ9YXyE81PdiDswoUWB_5e#scrollTo=M92yL8dH7sk3
# then load model here
```
