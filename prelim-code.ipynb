```python
model = YOLO("./models/yolo26x-seg.pt")  # load a pretrained model (recommended for training)
```

```python
results = model("testing-imgs/room0.png")
```

```python
# Access the results
for i, r in enumerate(results):
    # Plot results image
    im_bgr = r.plot()  # BGR-order numpy array
    im_rgb = Image.fromarray(im_bgr[..., ::-1])  # RGB-order PIL image

    # Show results to screen (in supported environments)
    r.show()

    # Save results to disk
    r.save(filename=f"./testing-results/results{i}.jpg")
```
