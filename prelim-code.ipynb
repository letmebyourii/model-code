# Importing/testing pre-trained detection model

Segmentation: 
https://docs.ultralytics.com/tasks/segment/#dataset-format

Hyperparameter tuning: https://pub.towardsai.net/understanding-hyper-parameter-tuning-of-yolos-82aec5f6e7b3


```python
model = YOLO("./models/yolo26n.pt")
```

```python
results = model("testing-imgs/room0.png",conf=0.05)
```

```python
# Access the results
for i, r in enumerate(results):
    # Plot results image
    im_bgr = r.plot()  # BGR-order numpy array
    im_rgb = Image.fromarray(im_bgr[..., ::-1])  # RGB-order PIL image

    # Show results to screen (in supported environments)
    r.show()

    # Save results to disk
    r.save(filename=f"./testing-results/results{i}.jpg")
```
